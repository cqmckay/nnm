This notebook recreates the procedure for finding nonlinear normal modes laid out in Peeters, et al.

```python
%pylab inline
import sympy
import numpy as np
import pandas as pd
from scipy.integrate import odeint
```

The first thing we need is the derivative function, so that we can do time integrations.  This is copied from the Double Pendulum notebook.

```python
def thetaddot(theta, t, params):
    theta1, theta2, theta1dot, theta2dot = theta
    
    m1 = params['m1']
    m2 = params['m2']
    l1 = params['l1']
    l2 = params['l2']
    g = params['g']
    
    M = np.matrix([[(m1+m2)*l1**2, m2*l1*l2*np.cos(theta1-theta2)],
                   [m2*l1*l2*np.cos(theta1-theta2), m2*l2**2]])
    B = np.matrix([[-m2*l1*l2*np.sin(theta1-theta2)*theta2dot**2 - (m1+m2)*g*l1*np.sin(theta1)],
                   [m2*l1*l2*np.sin(theta1-theta2)*theta1dot**2 - m2*g*l2*np.sin(theta2)]])
    
    resu = M.I*B
    theta1ddot = resu[0,0]
    theta2ddot = resu[1,0]
    return (theta1dot, theta2dot, theta1ddot, theta2ddot)
```

We will also want the functions for eigenfunctions and eigenvalues from that notebook, so that we have starting points for the continuation procedure.  As a reminder, we linearized the equations by hand, but then had sympy calculate the eigenvalues and eigenvectors. Then, we used `sympy.lambdify()` to convert those sympy expressions to a set of functions.

```python
m_1, m_2, l_1, l_2, g, omega = sympy.symbols("m_1, m_2, l_1, l_2, g, omega")
M = sympy.Matrix([[(m_1+m_2)*l_1**2, m_2*l_1*l_2],[m_2*l_1*l_2, m_2*l_2**2]])
B = sympy.Matrix([[-(m_1+m_2)*g*l_1, 0],[0, -m_2*g*l_2]])

Ematrix = (M.inv()*B)
Ematrix.simplify()
evects = Ematrix.eigenvects()

eval1 = sympy.lambdify((m_1, m_2, l_1, l_2, g), evects[0][0].simplify())
eval2 = sympy.lambdify((m_1, m_2, l_1, l_2, g), evects[1][0].simplify())
evect1 = sympy.lambdify((m_1, m_2, l_1, l_2, g), [evects[0][2][0][0], 1])
evect2 = sympy.lambdify((m_1, m_2, l_1, l_2, g), [evects[1][2][0][0], 1])
```

There are two pieces to the NNM calculation method outlined in Peeters, et al.: a shooting piece, and a continuation piece.  The shooting piece is used to refine a starting guess for the eigenmode, and the continuation piece is used to get the initial guess for the next energy. Let's start with the shooting piece, because it's a little easier to understand.

### Shooting

*Shooting* is a general method for solving boundary value problems.  In *initial value problems*, we have enough information to integrate the differential equation and find the solution.  In *boundary value problems*, we don't have sufficient information at the beginning of the interval, but we know where the solution is supposed to end up.  The basic procedure looks like this:

* Make a guess at the necessary parameters
* integrate the differential equations with those parameters (fire)
* check the boundary conditions (in our case, is it periodic?) (measure)
* refine the guess (aim)
* repeat until satisfied

In our specific case, we're looking for solutions which are periodic with a single, well-defined period. The parameters, in this case, are the initial conditions ${\mathbf{\Theta_0}}$ and the period $T$, and the boundary conditions are periodicity ($\mathbf\Theta(T) = \mathbf{\Theta_0}$, where $\mathbf\Theta = \left[\theta_1, \theta_2, \dot\theta_1, \dot\theta_2\right]^T$).

Firing is simple.

```python
def fire(theta0, T, returnfull=False):
    t = np.linspace(0, T, 500)
    thetaseries = odeint(thetaddot, theta0, t, args=(params,))
    if returnfull:
        return thetaseries
    else:
        return thetaseries[-1,:]
```

To see the problem, let's fire once with the linear eigenmode and eigenfrequency as guesses.

```python
m1 = 1.0
m2 = 1.0
l1 = 1.0
l2 = 1.0
g = 1.0

params = {'m1':m1,
          'm2':m2,
          'l1':l1,
          'l2':l2,
          'g':g}
T = 2*np.pi/np.sqrt(-eval1(1,1,1,1,1))
Theta0 = evect1(m1, m2, l1, l2, g)
Theta0.extend([0,0])

ts = fire(Theta0, T, returnfull=True)
```

If that's periodic, then $T$ is between half and 3/4 of a period.  However, it's not clear yet whether changing the period (on its own) is enough. It's possible that we're not on the right manifold. Let's proceed with measurement and aiming.

Measurement is also simple.

```python
def measure(theta0, thetaT):
    """Calculate the residual"""
    return thetaT - theta0

def converged(residual, tol):
    """Check convergence to within tol, using the L2 norm"""
    if sum(residual**2) <= tol:
        return True
    else:
        return False, sum(residual**2)
```

In our case,

```python
measure(Theta0, fire(Theta0, T))
```

We get four different values, since each of our four variables is off by a different amount.  We can see it isn't converged, but let's double check.

```python
converged(measure(Theta0, fire(Theta0, T)), 1.0e-16)
```

Aiming is where things get tricky. We're trying to get closer to four targets by adjusting five variables. That's four equations and five unknowns, which would be problematic, except that we can add another equation by setting the phase condition.  There are a lot of possibilities for this, but the one that comes to my mind as being easiest is picking $\dot\theta_1 = 0$.

The residual (the quantity given by our `measure` function) is supposed to be zero. Written out:
$${\mathbf H}({\mathbf \Theta}_{p0}, T) 􏰅\equiv {\mathbf \Theta}_{p}(T,{\mathbf \Theta}_{p0}) -􏰁 {\mathbf \Theta}_{p0} = {\mathbf 0}$$ The problem, then, is to find a zero crossing in the residual function. There are a number of algorithms we could apply, but the Newton-Raphson method has a nice balance of efficiency and stability, and it works in multiple dimensions, and doesn't have stringent demands for the initial guess.

The idea is that we approximate $\mathbf H$ by a linear function in each of its variables, and then extrapolate that linear function to zero. Since $\mathbf H$ is actually a nonlinear funciton, this won't get us all the way there, but it will provide our next guess. When the guesses get close enough, we stop.  Formally, we expand $\mathbf H$ in Taylor series (keeping only linear terms):
$${\mathbf H}({\mathbf \Theta}^{(0)}_{p0}, T^{(0)}) + {\partial {\mathbf H}\over \partial {\mathbf \Theta}_{p0}}\Delta {\mathbf \Theta}_{p0} + {\partial {\mathbf H}\over \partial T}\Delta T = 0,$$  And then solve for $\Delta {\mathbf\Theta}_{p0}$ and $\Delta T$.

If we write these equations all the way out (and include the phase condition), the method of solution becomes a little more clear:
$$
\begin{align}
{\partial H_1\over \partial\theta_1}\Delta\theta_1 + {\partial H_1\over \partial\theta_2}\Delta\theta_2 + {\partial H_1\over \partial\dot\theta_1}\Delta\dot\theta_1 + {\partial H_1\over \partial\dot\theta_2}\Delta\dot\theta_2 + {\partial H_1\over \partial T}\Delta T &= -H_1\\{\partial H_2\over \partial\theta_1}\Delta\theta_1 + {\partial H_2\over \partial\theta_2}\Delta\theta_2 + {\partial H_2\over \partial\dot\theta_1}\Delta\dot\theta_1 + {\partial H_2\over \partial\dot\theta_2}\Delta\dot\theta_2 + {\partial H_2\over \partial T}\Delta T &= -H_2\\{\partial H_3\over \partial\theta_1}\Delta\theta_1 + {\partial H_3\over \partial\theta_2}\Delta\theta_2 + {\partial H_3\over \partial\dot\theta_1}\Delta\dot\theta_1 + {\partial H_3\over \partial\dot\theta_2}\Delta\dot\theta_2 + {\partial H_3\over \partial T}\Delta T &= -H_3\\{\partial H_4\over \partial\theta_1}\Delta\theta_1 + {\partial H_4\over \partial\theta_2}\Delta\theta_2 + {\partial H_4\over \partial\dot\theta_1}\Delta\dot\theta_1 + {\partial H_4\over \partial\dot\theta_2}\Delta\dot\theta_2 + {\partial H_4\over \partial T}\Delta T &= -H_4\\
\dot\theta_1 &= 0
\end{align}
$$

Factoring
$$\left(\begin{array}{ccccc}
{\partial H_1\over \partial\theta_1}&{\partial H_1\over \partial\theta_2}&{\partial H_1\over \partial\dot\theta_1}&{\partial H_1\over \partial\dot\theta_2}&{\partial H_1\over \partial T}\\{\partial H_2\over \partial\theta_1}&{\partial H_2\over \partial\theta_2}&{\partial H_2\over \partial\dot\theta_1}&{\partial H_2\over \partial\dot\theta_2}&{\partial H_2\over \partial T}\\{\partial H_3\over \partial\theta_1}&{\partial H_3\over \partial\theta_2}&{\partial H_3\over \partial\dot\theta_1}&{\partial H_3\over \partial\dot\theta_2}&{\partial H_3\over \partial T}\\{\partial H_4\over \partial\theta_1}&{\partial H_4\over \partial\theta_2}&{\partial H_4\over \partial\dot\theta_1}&{\partial H_4\over \partial\dot\theta_2}&{\partial H_4\over \partial T}\\0&0&1&0&0
\end{array}\right)
\left(\begin{array}{c}\Delta\theta_1\\\Delta\theta_2\\\Delta\dot\theta_1\\\Delta\dot\theta_2\\\Delta T\end{array}\right) = -\left(\begin{array}{c}H_1\\H_2\\H_3\\H_4\\0\end{array}\right)
$$

So, the obvious thing to do is build the big matrix (which I'm going to call the augmented Jacobian), invert it, and apply it to the vector on the right hand side (which is just our residual augmented by a zero).  We can use finite difference approximations to build the augmented Jacobian one column at a time. The last column is more easily calculated by simply applying our derivative function to ${\mathbf\Theta}(T)$.

```python
def augjac(theta0, T):
    """Compute the augmented jacobian matrix and its inverse.
    
    I'm choosing clarity over efficiency here; if it becomes necessary to change that, I'll refactor."""
    
    # get the baseline residual
    final = fire(theta0, T)
    H0 = measure(theta0, final)
    aug0 = np.append(H0, 0.0)
    # choose a step size that is 0.01% of the residual
    steps = np.abs(H0*0.0001)
    
    jac = []

    # finite differences
    for idx, step in enumerate(steps):
        thetaidx0 = np.array(theta0)
        thetaidx0[idx] += step
        Hidx = measure(thetaidx0, fire(thetaidx0, T))
        col = (Hidx-H0)/step
        if idx == 2:
            col = np.append(col,[1.0],0)
        else:
            col = np.append(col,[0.0],0)
        jac.append(col)

    # last column is a little different
    dHdT = thetaddot(final, T, params)
    jac.append(np.append(np.array(dHdT), 0.0))
    
    #print "AUGJAC: ", np.matrix(np.array(jac)).T
    
    # turn it into a matrix and return the inverse
    return np.matrix(np.array(jac).T).I, np.matrix(aug0).T
```

```python
def getnewtheta(theta0, T):
    aj, ah = augjac(theta0, T)
    deltas = -aj * ah
#    print "NEWTHETA: ", ah, deltas
#    print "NEWTHETA: ", aj

#    print "NEWTHETA: ", aj.I*deltas
#    print "NEWTHETA: ", ah

    newtheta = np.append(theta0, T)
#    print newtheta
    return np.reshape(np.array(deltas),5) + np.reshape(newtheta,5)
```

Let's try this out. First, we'll check the periodicity of one of the linear eigenmodes at low energy.

```python
# get the in-phase linear eigenperiod
T = 2*np.pi/np.sqrt(-eval2(1,1,1,1,1))
# and the linear eigenmode for in-phase
Theta0 = evect2(m1, m2, l1, l2, g)
# add the \dot\theta terms
Theta0.extend([0,0])
# scale down to make the overall energy low
Theta0 = np.array(Theta0)*0.3
print "Guess:", T, Theta0
# integrate
ts = fire(Theta0, T, returnfull=True)
# is it converged?
print "Converged: ", converged(Theta0-ts[-1,:], 1.0e-8)
# actually, how close is it?
print measure(Theta0, ts[-1,:])
# and what does that look like, exactly?
pltres = plt.plot(np.linspace(0,T,500),ts)
```

Okay, so we have a guess which is close (it *looks* periodic) but not quite periodic. We need to start with a close guess because Newton-Raphson is a local optimization method, and will fail if we're too far away from the actual solution. Can the routines I've written above improve our guess?

```python
# get a new guess for theta0 and T.
newvec = getnewtheta(Theta0, T)
# separate into the chunks that fire expects
newT = newvec[-1]
newtheta = newvec[0:4]
print "New guess:", newT, newtheta
ts = fire(newtheta, newT, returnfull=True)
print "Converged: ", converged(newtheta-ts[-1,:], 1.0e-8)
print measure(newtheta, ts[-1,:])
pltres = plt.plot(np.linspace(0,newT,500),ts)
```
