# The double pendulum

## Introduction

```python
%pylab inline
import sympy
import numpy as np
import pandas as pd
from scipy.integrate import odeint
```

## The model

The two equations of motion are $$\left(m_1+m_2\right)l_1^2\ddot\theta_1 + m_2l_1l_2\cos\left(\theta_1-\theta_2\right)\ddot\theta_2 + m_2l_1l_2\sin\left(\theta_1-\theta_2\right)\dot\theta_2^2 + \left(m_1+m_2\right)gl_1\sin\theta_1 = 0$$
and
$$m_2l_1l_2\cos\left(\theta_1-\theta_2\right)\ddot\theta_1 + m_2l_2^2\ddot\theta_2  - m_2l_1l_2\sin\left(\theta_1-\theta_2\right)\dot\theta_1^2 + m_2gl_2\sin\theta_2 = 0.$$

Normally, we would separate the two equations into one for each of $\theta_1$ and $\theta_2$, but that's going to be a little messy in this case. Instead, let's note that at any particular instant in time, $\theta_1$ and $\theta_2$ have known values, and so we can write this as $$ M(\Theta(t)) \ddot\Theta(t) = B(\Theta(t)), $$ or equivalently, $$\ddot\Theta = M^{-1}B.$$

Here I'm using the notation $$\Theta = \left(\begin{array}{c} \theta_1 \\ \theta_2 \end{array} \right),$$ and have defined $$M = \left( \begin{array}{cc} \left(m_1+m_2\right)l_1^2 & m_2l_1l_2\cos\left(\theta_1-\theta_2\right) \\
m_2l_1l_2\cos\left(\theta_1-\theta_2\right) & m_2l_2^2\end{array}\right) $$ and $$ B = \left( \begin{array}{c} -m_2l_1l_2\sin\left(\theta_1-\theta_2\right)\dot\theta_2^2 - \left(m_1+m_2\right)gl_1\sin\theta_1 \\ m_2l_1l_2\sin\left(\theta_1-\theta_2\right)\dot\theta_1^2 - m_2gl_2\sin\theta_2 \end{array}\right). $$

Note that this is written in a nonstandard way because the equations are nonlinear. It would be more typical to have something of the form $M\ddot\Theta = B\Theta$, with $M$ and $B$ not depending on $\theta$. This means that we can't use standard linear algebra techniques on this (in particular, this is not a linear eigenvalue problem, and isn't really written in the form of an eigenvalue problem at all, though we could get there with a little massaging).


It would be possible to invert $M$ manually, but we might as well have the computer do it for us.  Hence our derivative function:

```python
def thetaddot(theta, t, params):
    theta1, theta2, theta1dot, theta2dot = theta
    
    m1 = params['m1']
    m2 = params['m2']
    l1 = params['l1']
    l2 = params['l2']
    g = params['g']
    
    M = np.matrix([[(m1+m2)*l1**2, m2*l1*l2*np.cos(theta1-theta2)],
                   [m2*l1*l2*np.cos(theta1-theta2), m2*l2**2]])
    B = np.matrix([[-m2*l1*l2*np.sin(theta1-theta2)*theta2dot**2 - (m1+m2)*g*l1*np.sin(theta1)],
                   [m2*l1*l2*np.sin(theta1-theta2)*theta1dot**2 - m2*g*l2*np.sin(theta2)]])
    
    resu = M.I*B
    theta1ddot = resu[0,0]
    theta2ddot = resu[1,0]
    return (theta1dot, theta2dot, theta1ddot, theta2ddot)
```

Let's try it out. For starters, set all of the lengths and masses to one, and start with the top pendulum hanging straight down, and the lower pendulum horizontal.

```python
thetazero = [0.0, np.pi/2, 0.0, 0.0]
params = {'m1':1.0,
          'm2':1.0,
          'l1':1.0,
          'l2':1.0,
          'g':1.0}
t = np.linspace(0, 100, 2000)
thetaseries = odeint(thetaddot, thetazero, t, args=(params,))
```

## Linearizing

We can linearize the above equations by taking the small angle approximation. This amounts to taking a Taylor series expansion and only keeping first order or lower terms in the small parameter (the angles).  If we stick with the nonstandard form I've used above, this leads us to new definitions for the parts of the matrix equation:
$$M = \left( \begin{array}{cc} \left(m_1+m_2\right)l_1^2 & m_2l_1l_2 \\
m_2l_1l_2 & m_2l_2^2\end{array}\right) $$ and 
$$ B = \left( \begin{array}{c} - \left(m_1+m_2\right)gl_1\theta_1 \\  - m_2gl_2\theta_2 \end{array}\right). $$

Let's implement it.

```python
def thetaddot_linear(theta, t, params):
    theta1, theta2, theta1dot, theta2dot = theta
    
    m1 = params['m1']
    m2 = params['m2']
    l1 = params['l1']
    l2 = params['l2']
    g = params['g']
    
    M = np.matrix([[(m1+m2)*l1**2, m2*l1*l2],
                   [m2*l1*l2, m2*l2**2]])
    B = np.matrix([[ - (m1+m2)*g*l1*theta1],
                   [ - m2*g*l2*theta2]])
    
    resu = M.I*B
    theta1ddot = resu[0,0]
    theta2ddot = resu[1,0]
    return (theta1dot, theta2dot, theta1ddot, theta2ddot)
```

And a test...

```python
thetazero = [0.0, np.pi/2, 0.0, 0.0]
params = {'m1':1.0,
          'm2':1.0,
          'l1':1.0,
          'l2':1.0,
          'g':1.0}
t = np.linspace(0, 100, 2000)
thetaseries_lin = odeint(thetaddot_linear, thetazero, t, args=(params,))
tpd_lin = pd.DataFrame(thetaseries_lin, index=t, columns=[r"$\theta_1$", r"$\theta_2$", r"$\dot\theta_1$", r"$\dot\theta_2$"])
```

## Energetics

We derived the equations of motion by using the Lagrangian, which is made up of a combination of the kinetic and potential energies.  Let's look at each of those now.

The potential energy is all gravitational, and has one term for each of the coordinates:
$$U = (m_1+m_2) g l_1 \left(1-\cos\theta_1\right) + m_2 g l_2\left(1-\cos\theta_2\right)$$  The key thing to note here is that changing $\theta_1$ moves both masses, while changing $\theta_2$ only moves the lower mass, which is why the two terms look a little different.  It also means that even if the two masses are the same, these two energies will scale differently.

The kinetic energy is more interesting.  Again, changing $\theta_1$ will move both masses, so that effect must be included. Additionally, however, the position of $\theta_1$ determines the origin of the coordinate system for $\theta_2$, which gives rise to an third term in the kinetic energy:
$$T = {1\over2}\left(m_1+m_2\right)l_1^2\dot\theta_1^2 + {1\over2}m_2l_2^2\dot\theta_2^2 + m_2l_1l_2\cos\left(\theta_1-\theta_2\right)\dot\theta_1\dot\theta_2.$$
The cosine appears here as a result of the fact that we're taking the dot product of the velocity vectors of the two masses.

Expressed in this way, we see 5 distinct reservoirs of energy: 
1. Potential energy in the top pendulum ($U_1 = (m_1+m_2) g l_1 \left(1-\cos\theta_1\right)$)
2. Potential energy in the bottom pendulum ($U_2 = m_2 g l_2\left(1-\cos\theta_2\right)$)
3. Kinetic energy in the top pendulum ($T_1 = {1\over2}\left(m_1+m_2\right)l_1^2\dot\theta_1^2 $)
4. Kinetic energy in the bottom pendulum ($T_2 = {1\over2}m_2l_2^2\dot\theta_2^2$)
5. "Bending" kinetic energy ($T_B = m_2l_1l_2\cos\left(\theta_1-\theta_2\right)\dot\theta_1\dot\theta_2$)

Having figured out what kinds of energy we have at play in our system, we have two general questions. First, how much energy is in each reservoir (the question of *Energy Partition*). Second, how does energy move between reservoirs (the question of *Energy Transfer*).

#### Energy Partition

Our first guess might be that energy is divided equally among the five categories. Now, obviously, this can't be true at every instant, but it might be true in a time-averaged sense.  Let's check this out for our time series.  While we're at it, we'll double check that total energy is conserved.

We calculate each of the energy terms, and then add them to get total potential and kinetic energies, and then add those to get total energy.

```python
tpd[r"$U_1$"] = (params['m1'] + params['m2'])*params['g']*params['l1']*(1-np.cos(tpd[r"$\theta_1$"]))
tpd[r"$U_2$"] = params['m2']*params['g']*params['l2']*(1-np.cos(tpd[r"$\theta_2$"]))
tpd[r"$T_1$"] = 0.5*(params['m1'] + params['m2'])*params['l1']**2*tpd[r"$\dot\theta_1$"]**2
tpd[r"$T_2$"] = 0.5*params['m2']*params['l2']**2*tpd[r"$\dot\theta_2$"]**2
tpd[r"$T_B$"] = params['m2']*params['l1'] * params['l2']*np.cos(tpd[r"$\theta_1$"]-tpd[r"$\theta_2$"])*tpd[r"$\dot\theta_1$"]*tpd[r"$\dot\theta_2$"]
tpd[r"$U_{tot}$"] = tpd[r"$U_1$"] + tpd[r"$U_2$"]
tpd[r"$T_{tot}$"] = tpd[r"$T_1$"] + tpd[r"$T_2$"] + tpd[r"$T_B$"]
tpd[r"$E_{tot}$"] = tpd[r"$U_{tot}$"] + tpd[r"$T_{tot}$"]
energies = [r"$U_1$", r"$U_2$", r"$T_1$", r"$T_2$", r"$T_B$", r"$T_{tot}$", r"$U_{tot}$", r"$E_{tot}$"]
tpd[energies].describe()
```

Good news! Our model conserves energy. As far as equipartition goes, though, there's something funny going on. Setting aside the variation between $U_1$ and $U_2$, or $T_1$ and $T_2$, or even the kinetic and potential energies, the "Bending" kinetic energy is clearly an oddball. First of all, it's negagive most of the time, which seems like an odd situation for a kinetic energy.

We need to be careful, though. This particular energy isn't defined as a velocity squared, but rather as a dot product of two different velocities. As such, if the two pendula are moving in opposite directions (and the angle between them is less than $\pi/2$), this energy will be negative. Essentially that says that for a given amount of kinetic energy, the two angles can change faster if they're moving in opposite directions than if they're moving in the same direction.

Just for kicks, let's plot the energies and see what they look like.

```python
ax = tpd[[r"$U_1$", r"$U_2$", r"$T_1$", r"$T_2$", r"$T_B$"]].plot(linewidth=2)
fig = ax.get_figure()
fig.set_figwidth(10)
fig.set_figheight(3)
ax.set_xlim(0,21)
leg = plt.legend(loc='upper left')
leg.set_bbox_to_anchor((1.0, 1.0))
```

```python
ax = tpd.plot(x=r"$U_2$", y=r"$T_B$")
```

The numbers are suggestive of the idea that the total energy is evenly split between potential and kinetic.  Also, if we have four positive reservoirs and one negative, putting a third of the total energy in each would be equal balance, and the numbers for each reservoir are roughly in that neighborhood.

If the deviations from these guesses are due to statistical variations, we should be able to push the means toward their target values by conducting longer runs.  It isn't as simple as increasing the number of samples, because these are time averages; it's actually the number of oscillations that is important, rather than the number of samples.  We can check this out by trying a different time interval.

```python
def energy_test(ntime, nsamples):
    thetazero = [0.0, np.pi/2, 0.0, 0.0]
    params = {'m1':1.0,
          'm2':1.0,
          'l1':1.0,
          'l2':1.0,
          'g':1.0}
    t = np.linspace(0, ntime, nsamples)
    thetaseries = odeint(thetaddot, thetazero, t, args=(params,))
    tpd = pd.DataFrame(thetaseries, index=t, columns=[r"$\theta_1$", r"$\theta_2$", r"$\dot\theta_1$", r"$\dot\theta_2$"])
    tpd[r"$U_1$"] = (params['m1'] + params['m2'])*params['g']*params['l1']*(1-np.cos(tpd[r"$\theta_1$"]))
    tpd[r"$U_2$"] = params['m2']*params['g']*params['l2']*(1-np.cos(tpd[r"$\theta_2$"]))
    tpd[r"$T_1$"] = 0.5*(params['m1'] + params['m2'])*params['l1']**2*tpd[r"$\dot\theta_1$"]**2
    tpd[r"$T_2$"] = 0.5*params['m2']*params['l2']**2*tpd[r"$\dot\theta_2$"]**2
    tpd[r"$T_B$"] = params['m2']*params['l1'] * params['l2']*np.cos(tpd[r"$\theta_1$"]-tpd[r"$\theta_2$"])*tpd[r"$\dot\theta_1$"]*tpd[r"$\dot\theta_2$"]
    tpd[r"$U_{tot}$"] = tpd[r"$U_1$"] + tpd[r"$U_2$"]
    tpd[r"$T_{tot}$"] = tpd[r"$T_1$"] + tpd[r"$T_2$"] + tpd[r"$T_B$"]
    tpd[r"$E_{tot}$"] = tpd[r"$U_{tot}$"] + tpd[r"$T_{tot}$"]
    energies = [r"$U_1$", r"$U_2$", r"$T_1$", r"$T_2$", r"$T_B$", r"$T_{tot}$", r"$U_{tot}$", r"$E_{tot}$"]
    return tpd[energies].describe()
```

If the differences between the energies were purely statistical, I would expect them to be reduced by about a factor of 3 (actually, $\sqrt{10}$) when I increase the time interval by a factor of 10, but we don't observe that. So, I conclude that the differences between the energies are real. I don't feel like I understand that yet, exactly, but it's time to move on to energy transfer.

#### Energy transfer

Taking the time derivative of our expressions for $U$ and $T$ gives us some energy evolution equations. We don't evolve these directly, but looking at the terms allows us to see what the contributing factors are for each of our energy buckets.  For potential energy, we have 
$$\dot U = (m_1+m_2) g l_1 \dot\theta_1\sin\theta_1 + m_2 g l_2\dot\theta_2\sin\theta_2,$$ 
while kinetic energy gives us 
$$\dot T = \left(m_1+m_2\right)l_1^2\dot\theta_1\ddot\theta_1 + m_2l_2^2\dot\theta_2\ddot\theta_2 + \left[  m_2l_1l_2\cos\left(\theta_1-\theta_2\right)\ddot\theta_1\dot\theta_2 
+ m_2l_1l_2\cos\left(\theta_1-\theta_2\right)\dot\theta_1\ddot\theta_2 
- m_2l_1l_2\sin\left(\theta_1-\theta_2\right)\dot\theta_1^2\dot\theta_2 
+ m_2l_1l_2\sin\left(\theta_1-\theta_2\right)\dot\theta_1\dot\theta_2^2 
\right].$$ 
The top and bottom layer energies are easily identifiable, and the expression in square brackets is the time derivative of the bending kinetic energy.

If we multiply the first of our model equations by $\dot\theta_1$, and the second by $\dot\theta_2$, we see how the energies are coupled together:

$$\left(m_1+m_2\right)l_1^2\dot\theta_1\ddot\theta_1 + m_2l_1l_2\cos\left(\theta_1-\theta_2\right)\dot\theta_1\ddot\theta_2 + m_2l_1l_2\sin\left(\theta_1-\theta_2\right)\dot\theta_1\dot\theta_2^2 + \left(m_1+m_2\right)gl_1\dot\theta_1\sin\theta_1 = 0$$
and
$$m_2l_1l_2\cos\left(\theta_1-\theta_2\right)\dot\theta_2\ddot\theta_1 + m_2l_2^2\dot\theta_2\ddot\theta_2  - m_2l_1l_2\sin\left(\theta_1-\theta_2\right)\dot\theta_2\dot\theta_1^2 + m_2gl_2\dot\theta_2\sin\theta_2 = 0.$$

We can see that the first equation contains $\dot T_1$, $\dot U_1$, and two of the terms of $\dot T_B$, while the other equation has $\dot T_2$, $\dot U_2$, and the other two terms of $\dot T_B$.

## Normal Modes

With the equations linearized, we're able to switch back to the more common (and perhaps more useful) form of a linear eigenvalue problem:
$$M\ddot\Theta = B\Theta,$$
where
$$M= \left( \begin{array}{cc} \left(m_1+m_2\right)l_1^2 & m_2l_1l_2 \\
m_2l_1l_2 & m_2l_2^2\end{array}\right) $$ as above, and
$$ B = \left( \begin{array}{cc} - \left(m_1+m_2\right)gl_1 & 0\\0&  - m_2gl_2 \end{array}\right). $$

Actually, I lied a little bit. This still isn't *quite* in the form of an eigenvalue problem yet. We need to make one more assumption, and then rearrange. The assumption is this: the system will oscillate with a single frequency. In that case, $$\ddot\Theta \rightarrow -\omega^2\Theta, $$ and we can rearrange the matrix equation to be $$M^{-1}B\Theta + \omega^2\Theta = 0,$$ or, identifying the eigenvalue $\lambda = -\omega^2$, $$\left(M^{-1}B - \lambda I\right)\Theta = 0.$$
