I'm debugging shooting, and while I think I know what's going on, I want to try it with a simpler system to double check.  So, I'm going to do the same thing with a single pendulum. In this case, since there isn't any mode coupling (just one eigenmode), there's no chaos, so even at high energies we should be able to find a stable solution.

```python
%pylab inline
import sympy
import numpy as np
import pandas as pd
from scipy.integrate import odeint
```

```python
def zdot(z, t, params):
    x1, x1dot = z
    
    m1 = params['m1']
    l1 = params['l1']
    g = params['g']
    
    x1ddot = -g/l1 * np.sin(x1)
    return (x1dot, x1ddot)
```

```python
def fire(z0, T, returnfull=False):
    t = np.linspace(0, T, 500)
    thetaseries = odeint(zdot, z0, t, args=(params,))
    if returnfull:
        return thetaseries
    else:
        return thetaseries[-1,:]
```

```python
def measure(theta0, thetaT):
    """Calculate the residual"""
    return thetaT - theta0

def converged(residual, tol):
    """Check convergence to within tol, using the L2 norm"""
    if sum(residual**2) <= tol:
        return True
    else:
        return False, sum(residual**2)
```

```python
def augjac(theta0, T):
    # get the baseline residual
    final = fire(theta0, T)
    H0 = measure(theta0, final)
    #print "augjac: H0", H0
    aug0 = np.append(H0, 0.0)
    # choose a step size that is 0.01% of the residual
    steps = np.abs(H0*0.0001)
    
    jac = []

    # finite differences
    for idx, step in enumerate(steps):
        #print "augjac: idx, step", idx, step
        thetaidx0 = np.array(theta0)
        #print "augjac: theta0", thetaidx0
        thetaidx0[idx] += step
        #print "augjac: shifted", thetaidx0
        Hidx = measure(thetaidx0, fire(thetaidx0, T))
        #print "augjac: Hidx", Hidx
        #print "augjac: diff", Hidx-H0
        col = (Hidx-H0)/step
        #print "augjac: col", col
        if idx == 1:
            col = np.append(col,[1.0],0)
        else:
            col = np.append(col,[0.0],0)
        jac.append(col)

    # last column is a little different
    dHdT = zdot(final, T, params)
    jac.append(np.append(np.array(dHdT), 0.0))
    
    #print "AUGJAC: ", np.matrix(np.array(jac)).T
    
    # turn it into a matrix and return the inverse
    return np.matrix(np.array(jac).T).I, np.matrix(aug0).T
```

```python
def getnewtheta(theta0, T):
    aj, ah = augjac(theta0, T)
    deltas = -aj * ah
    #print "NEWTHETA: ", ah
    #print "deltas: ", deltas
    #print "NEWTHETA: ", aj

#    print "NEWTHETA: ", aj.I*deltas
#    print "NEWTHETA: ", ah

    newtheta = np.append(theta0, T)
#    print "NEWTHETA", newtheta
#    print "NEWTHETA", deltas
    return np.reshape(np.array(deltas),3) + np.reshape(newtheta,3)
```

In the linear case, $T=2\pi\sqrt{l\over g}$. We can use this as our approximation (or initial guess).  If the amplitude is small enough, the shift in frequency is quite small, and we're converged as we start.

```python
z0 = np.array((0.01, 0.0))
T = 2*np.pi
params = {'g':1.0, 'l1':1.0, 'm1':1.0}

ts = fire(z0, T, returnfull=True)
# is it converged?
print "Converged: ", converged(z0-ts[-1,:], 1.0e-8)
# actually, how close is it?
print measure(z0, ts[-1,:])
# and what does that look like, exactly?
pltres = plt.plot(np.linspace(0,T,500),ts)
plt.grid()
```

The whole point of Newton's method is to turn an unconverged result into a converged result. We do this by (essentially) expanding the function in Taylor Series, and then solving for the necessary correction. Since we're only keeping the linear terms of the Taylor Series, we don't actually get our solution in one step; it usually takes a few iterations. However, as long as we start close enough to an actual solution, we get there pretty quickly. It's still kind of a pain to do by hand, so I'll put it in a function.

```python
def shoot(guess, tol):
    notconv = True
    z0, T0 = guess
    
    while notconv:
        zT = fire(z0, T0, returnfull=False)
        l2norm = sum(measure(z0,zT)**2)
        if l2norm >= tol:
            # for debugging purposes, can we watch it converge?
            print l2norm
            notconv = True
            newvec = getnewtheta(z0, T0)
            T0 = newvec[-1]
            z0 = newvec[0:2]
        else:
            # mostly because I'm curious...
            print "Converged.", l2norm
            notconv = False
        
    return z0, T0
```
